模块4：数据可视化Plot
1. 简介
Pandas是一个强大的数据处理库，其中plot模块提供了简单易用的数据可视化功能，可以帮助用户快速生成各种类型的图表。
2. Plot 模块的功能
包括图表标题、坐标轴标签、图例、颜色、线型等。同时它是基于Matplotlib 库构建的，因此用户可以通过 plot 方法生成的图表对象进一步进行 Matplotlib 的定制化操作，实现更高级的定制和可视化效果。总结Plot有以下几个功能特点：多种图表类型，简单方便，支持自定义，数据处理和可视化一体化。
以下为几个常用的功能：
生成折线图
data.plot(kind='line')
生成条形图
df.plot(kind='bar')
生成散点图
data.plot(kind='scatter', x='A', y='B')
生成直方图
data.plot(kind='hist')
生成箱线图
data.plot(kind='box')
自动调整子图参数以使其适合图形区域
tight_layout
在一张大画布上创建多个小图
Subplot
指定生成图表的类型、横纵坐标、标题、颜色、样式等信息
df.plot(x='x', y='y', kind='line', title='Line Plot', color='blue')

可视化示例：

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    'A': [1, 2, 1, 3, 3, 2, 4, 1],
    'B': [4, 3, 2, 1, 4, 3, 2, 1]
})
```

1.绘制直方图

```python
data.plot(kind='hist', alpha=0.7)
plt.show()
```

2.折线图

```python
data = pd.Series([1, 2, 3, 4, 5])
data.plot(kind='line')
plt.show()
```

3.条形图

```python
data.plot(kind='bar')
plt.show()
```

4.散点图

```python
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1]
})
data.plot(kind='scatter', x='A', y='B')
plt.show()
```

5.箱型图

```python
data.plot(kind='box')
plt.show()
```

6.多个小图

```python
x = [1, 2, 3, 4, 5]
y1 = [5, 4, 3, 2, 1]
y2 = [1, 2, 3, 4, 5]
fig, axs = plt.subplots(2)

axs[0].plot(x, y1)

axs[1].plot(x, y2)

plt.show()
```

7.画饼图

```python
data = {'sizes': [30, 20, 15, 35],
        'labels': ['A', 'B', 'C', 'D']}
df = pd.DataFrame(data)
fig, ax = plt.subplots()
df.plot(kind='pie', y='sizes', labels=df['labels'], autopct='%1.1f%%', startangle=90, ax=ax, legend=False)
ax.set_title('Pie Chart Example')
plt.axis('equal')  # 让饼图呈正圆形
plt.show()
```
